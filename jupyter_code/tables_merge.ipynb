```python
import pandas as pd
import numpy as np
```

```python
one_bedroom = pd.read_csv('zillow_1bedroom_transformed.csv')
two_bedroom = pd.read_csv('zillow_2bedroom_transformed.csv')
three_bedroom = pd.read_csv('zillow_3bedroom_transformed.csv')
four_bedroom = pd.read_csv('zillow_4bedroom_transformed.csv')
five_bedroom = pd.read_csv('zillow_5bedroom_transformed.csv')
```

```python
five_bedroom.head(5)
```

```python
merged = pd.concat([one_bedroom, two_bedroom, three_bedroom, four_bedroom, five_bedroom], ignore_index=True)
```

```python
df = merged.sort_values(['ZipCode', 'Bedrooms'])
```

```python
df = df.drop(columns='Unnamed: 0')
```
