#### We need to split the original data into smaller tables to query in SQL. As it stands each year is it 's own columns which makes it very difficult ato analyze home prices by year. Specifically here we are unpivoting the dataset so we can get a column called 'Year' whcih can be used for analysis in SQL.

```python
df.drop(columns=['State', 'County', 'City', 'ZipCode'])
```

```python
# Identifying the year columns for unpivoting
year_cols = [str(year) for year in range(2000,2024)]

# Identifier columns are 'RegionID' and 'Bedrooms'
identifier_cols = ['RegionID', 'Bedrooms']

# Using pandas melt to unpivot the data
unpivoted_dataset = pd.melt(df, id_vars = identifier_cols, value_vars = year_cols, var_name = 'Year', value_name = 'Price')
```
