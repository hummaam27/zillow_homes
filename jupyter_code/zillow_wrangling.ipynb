# Data preparation

1. Delete redundant and irrelevant rows
2. Rearrange columns for better redability
3. Add column with data relating do amount of bedrooms



```python
# A check to make sure columns we are dropping are actually redundant.
same_values = (df['StateName'] == df['State']).all()
print(same_values)
```

```python
#drop extra and irrelevant columns for out project
df.drop(columns = ['SizeRank', 'RegionType', 'StateName', 'Metro'], inplace = True)
```

```python
df.head(5)
```

```python
# Rename some columns.
df.rename(columns = {'RegionName': 'ZipCode', 'CountyName': 'County'}, inplace = True)
```

```python
# Reorder location columns to make the location heirarchy better
cols = df.columns.tolist()
cols[1], cols[2] = cols[2], cols[1]
df =df[cols]
```

```python
cols[2], cols[4] = cols[4], cols[2]
df =df[cols]
```

```python
# Add a column for the number of bedrooms. For this project there is seperate datasets for bedrooms 1-5 and we nned to input this info.
df.insert(5, 'Bedrooms', 5)
```

```python
df.head(3)
```

```python
#the date columns are in months and are over 200. In this project we won't need this granularity so we can group the dates by year and thereby simplify the dataset.
date_columns = df.columns[6:]
year_mapping = {col: pd.to_datetime(col).year for col in date_columns}
grouped_by_year = df[date_columns].T.groupby(year_mapping).mean().T
grouped_by_year_round = grouped_by_year.round(2)
non_date_columns = df.iloc[:, :6]
simplified_df = pd.concat([non_date_columns, grouped_by_year_round], axis=1)
```
